```python
import pandas as pd
```

```python
df = pd.read_csv('./assignment data.csv')
```

```python
df.head()
```

```python
#check null values in columns
df.isnull().sum()
```

```python
# calculate price per square foot and average price per square foot
df['price_per_sqft'] = df['price'] / df['sq__ft']
```

```python
average_price_per_foot = df['price_per_sqft'].mean()
```

```python
# filter data for properties sold for less than the average price per square foot
filtered_data = df[df['price_per_sqft'] < average_price_per_foot]

# Write the filtered data to a new CSV file
filtered_data.to_csv('filtered-sales-data.csv', index=False)
```
